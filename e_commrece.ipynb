```python
missing_data = df.isnull().sum() / len(df) * 100
print(missing_data[missing_data > 0])  # Show only columns with missing values
```

```python
df = df.drop(columns=missing_data[missing_data > 70].index)
```

```python
df.columns
```

```python
df.head()
```

```python
missing_data = df.isnull().sum() / len(df) * 100
print(missing_data[missing_data > 0])  # Show only columns with missing values
```

```python
df = df.dropna()
```

```python
df['created_at']=pd.to_datetime(df['created_at'])
df['Working Date']=pd.to_datetime(df['Working Date'])
df['M-Y']=pd.to_datetime(df['M-Y'])
```

```python
df['status'].fillna('No Idea')
df['sku'].fillna('Unknown')
```

```python
df['qty_ordered'] = df['qty_ordered'].fillna(df['qty_ordered'].mean())  # Mean
df['price'] = df['price'].fillna(df['price'].median())  # Median
df['grand_total'] = df['grand_total'].fillna(df['grand_total'].median())
```

```python
df.isnull().sum()
```

```python
sns.heatmap(df.isnull())
plt.title("Missing Data Heatmap")
plt.show()
```

```python
df['month'] = df['created_at'].dt.to_period('M')
monthly_sales = df.groupby('Month')['price'].sum()
monthly_sales.plot(kind='line', figsize=(10, 6), title='Monthly Sales Trends')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.show()
```

```python
top_categories = df['category_name_1'].value_counts()
top_categories.plot(kind='bar', figsize=(10, 6), title='Top Selling Categories')
plt.xlabel('category_name_1')
plt.ylabel('qty_ordered')
plt.show()
```

```python
revenue_by_category = df.groupby('category_name_1')['price'].sum()
revenue_by_category.plot(kind='bar', figsize=(10, 6), title='Revenue by Category')
plt.xlabel('Category')
plt.ylabel('Total Revenue')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
df['payment_method'].value_counts().plot(kind='bar', color='skyblue')  # Use 'bar' for a bar chart
plt.title('Payment Method Distribution', fontsize=16)
plt.xlabel('Payment Methods', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
category_status_counts = df.groupby(['category_name_1', 'status']).size().reset_index(name='counts')

# Step 2: Create the bar chart with Seaborn
plt.figure(figsize=(16, 10))
sns.barplot(
    data=category_status_counts,
    x='category_name_1', 
    y='counts', 
    hue='status', 
)

# Step 3: Add title and labels
plt.title('Order Categories with Status Breakdown', fontsize=16)
plt.xlabel('Categories', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.legend(title='Order Status', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=1)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Aggregate the data and filter top 10 categories
top_categories = df['category_name_1'].value_counts().head(10).index
filtered_data = df[df['category_name_1'].isin(top_categories)]
category_status_counts = filtered_data.groupby(['category_name_1', 'status']).size().reset_index(name='counts')

# Step 2: Create the bar chart
plt.figure(figsize=(14, 8))
sns.barplot(
    data=category_status_counts,
    x='category_name_1', 
    y='counts', 
    hue='status',
)

# Step 3: Customize the chart
plt.title('Top 10 Categories with Status Breakdown', fontsize=16)
plt.xlabel('Categories', fontsize=14)
plt.ylabel('Counts', fontsize=14)
plt.xticks(rotation=45)
plt.legend(title='Order Status', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
